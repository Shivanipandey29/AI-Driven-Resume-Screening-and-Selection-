```python
lst_skills=[]
d={}
for i in df["Skills/Description"]:
    for j in i.split(","):
        lst_skills.append(j)

        
for i in list_skills:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
        
for w in sorted(d, key=d.get, reverse=True):
    print(w, d[w])
```

```python
lst_skills = []
d = {}

# Assuming df["Skills/Description"] contains the skills and descriptions
for i in df_new["Skills/Description"]:
    for j in i.split(","):
        lst_skills.append(j.lower())  # Convert to lowercase

for skill in lst_skills:
    if skill in d:
        d[skill] += 1
    else:
        d[skill] = 1

for w in sorted(d, key=d.get, reverse=True):
    print(w, d[w])
```
